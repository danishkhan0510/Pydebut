```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
import warnings
warnings.filterwarnings(action="ignore")
```

```python
auto = pd.read_csv("/content/Automobile (1).csv")
```

```python
auto.head()
```

```python
auto.describe(include="all")
```

```python
auto.info()
#there are no null values
```

```python
#Histogram
sns.distplot(auto['normalized_losses'],kde=False,rug=True)
plt.show()
```

```python
#Scatter_Plot
sns.jointplot(auto['engine_size'],auto['horsepower'])
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower'],kind="hex")
```

```python
sns.jointplot(auto['engine_size'],auto['horsepower'],kind="kde")
```

```python
#Paired_Plots
sns.pairplot(auto[['engine_size','height','length']])
```

```python
#Strip_Plot
sns.stripplot(auto['engine_type'],auto['length'],jitter=True)
```

```python
#Swarm_Plot
sns.swarmplot(auto['engine_type'],auto['length'])
```

```python
#Box_Plot
sns.boxplot(auto['horsepower'],auto['fuel_type'],hue=auto['number_of_doors'])
```

```python
#Bar_Plot
sns.barplot(auto['engine_type'],auto['horsepower'],hue=auto['number_of_doors'])
```

```python
#Count_Plot
sns.countplot(auto['number_of_cylinders'],hue=auto['number_of_doors'])
```

```python
#Point_Plot
sns.pointplot(auto['engine_type'],auto['horsepower'],hue=auto['fuel_type'])
```

```python
#Factor_Plot
sns.factorplot(x='engine_type',y='horsepower',hue='number_of_doors',col='fuel_type',data=auto,kind='swarm')
```

```python
#Linear_Regression
sns.lmplot(x="horsepower",y="length",data=auto,hue="fuel_type")
```
